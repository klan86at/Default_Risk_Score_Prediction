```python
# Importing Necessary Libraries for the test data
import pickle
import pyforest
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error
from sklearn.metrics import mean_squared_error, r2_score, root_mean_squared_error
import warnings
warnings.filterwarnings("ignore")  # Suppresses all warnings
```

```python
# Load the trained model
with open("gb_best.pkl", "rb") as file:
    gb_best = pickle.load(file)
```

```python
# Read the test data into a dataframe
df_test = pd.read_csv(r"D:\LuxDev\ClassModels\test.csv")
df_test.head()
```

```python
# Split(vertical) the data
target = "default_risk_score"
X = df_test.drop(columns=[target])
y = df_test[target]
```

```python
# Drop columns that were dropped in the train dataset
X = X.drop(columns=['credit_score'], axis=1)
X.head()
```

```python
# Scale our data
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
# We do predictions using our generated models
test_pred = gb_best.predict(X)
mse = mean_absolute_error(y, test_pred)
mape = mean_absolute_percentage_error(y, test_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y, test_pred)
print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"MAPE: {mape:.2f}")
print(f"R2 Score: {r2:.2f}")
```

```python
# we add to our test column a new col pred_default_risk_gb
df_test['pred_default_risk_gb'] = pd.DataFrame(test_pred)
df_test.head()
```
