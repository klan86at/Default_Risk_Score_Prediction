```python
# Load the trained model
with open("gb_best.pkl", "rb") as file:
    gb_best = pickle.load(file)
```

```python
df_test= pd.read_csv(r"D:\LuxDev\ClassModels\test.csv")
df_test.head()
```

```python
#standard scale the test data
scaler= StandardScaler()
scaled_feat= ['loan_term', 'interest_rate', 'debt_to_income_ratio', 'employment_years', 'age']
df_test[scaled_feat]= scaler.fit_transform(df_test[scaled_feat])
#Log transfrorm credit_score, income, loan amount, savings balance & default risk score for the test data
log_feat= ['credit_score', 'income', 'loan_amount', 'savings_balance', 'default_risk_score']
df_test[log_feat]= np.log1p(df_test[log_feat])
```

```python
#Split(vertical) the data
target= "default_risk_score"
X= df_test.drop(columns=[target])
y= df_test[target]
```

```python
X= X.drop(columns= ['credit_score', 'income', 'savings_balance' ], axis=1)
X.head()
```

```python
test_pred= gb_best.predict(X)
mse= mean_absolute_error(y, test_pred)
mape= mean_absolute_percentage_error(y, test_pred)
rmse = np.sqrt(mse)
r2= r2_score(y, test_pred)
print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"MAPE: {mape:.2f}")
print(f"R2 Score: {r2:.2f}")
```
